```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'serif'


# Load psi data from a CSV file
psi = np.loadtxt('psi.csv', delimiter=',')

# Define grid and spacing
dx = 1/1024
dy = dx
x = np.linspace(0, 1, 1024)
y = np.linspace(0, 1, 1024)

# Calculate Ex and Ey
Ex = -(np.roll(psi, -1, axis=1) - np.roll(psi, 1, axis=1)) / (2.0 * dx)
Ey = -(np.roll(psi, -1, axis=0) - np.roll(psi, 1, axis=0)) / (2.0 * dy)

# Create a figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot psi using imshow
im = ax.imshow(psi, cmap='Spectral', extent=[0, 1, 0, 1], origin='lower')

ax.streamplot(x, y, Ex, Ey, color = 'black')
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Electric Field Strength')

# Customize axis labels and ticks
ax.set_title('V(x,y) with electric field lines overlayed')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.savefig("problem1.png")
plt.show()
```
